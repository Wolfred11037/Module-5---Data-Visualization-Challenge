```python
# Create Bar Chart DataFrame.plot

# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
combined_data = pd.merge(mouse_metadata, study_results, on="Mouse ID", how="outer")

# Generate bar plot for each drug regimen
DataFrame = combined_data["Drug Regimen"].value_counts()

# implementing DataFrame.plot
DataFrame.plot(kind="bar", figsize=(10, 6))
plt.xlabel("Drug Regimen")
plt.ylabel("Number of Rows")
plt.title("Total Number of Rows for Each Drug Regimen")
plt.xticks(rotation=45)
plt.show()
```

```python
# Create Bar Chart pyplot

# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
combined_data = pd.merge(mouse_metadata, study_results, on="Mouse ID", how="outer")

# Calculating total number of rowsfor each drug regimen
pyplot = combined_data["Drug Regimen"].value_counts()

# Generate bar plot via pyplot
plt.figure(figsize=(10, 6))
plt.bar(pyplot.index, pyplot.values)
plt.xlabel("Drug Regimen")
plt.ylabel("Number of Rows")
plt.title("Total Number of Rows for Each Drug Regimen")
plt.xticks(rotation=45)
plt.show()
```

```python
# Create Pie Chart DataFrame.plot()

# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
combined_data = pd.merge(mouse_metadata, study_results, on="Mouse ID", how="outer")

# Calculation between female versus male mice
gender_sorting = combined_data["Sex"].value_counts()

# Pie plot via Pandas DataFrame.plot() method
gender_sorting.plot(kind="pie", autopct="%1.1f%%", startangle=45)
plt.axis("equal")
plt.title("Sex")
plt.show()
```

```python
# Create Pie Chart pyplot

# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
combined_data = pd.merge(mouse_metadata, study_results, on="Mouse ID", how="outer")

# Calculation between female versus male mice
gender_distribution = combined_data["Sex"].value_counts()

# Pie plot via pyplot
plt.figure(figsize=(6, 6))
plt.pie(gender_distribution, labels=gender_distribution.index, autopct="%1.1f%%", startangle=45)
plt.axis("equal") 
plt.title("Sex")
plt.show()
```
